```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open(r'../../results/outcomes/S1-under-proportion.csv') as S1_under_prop:
    df_S1_under_prop=pd.read_csv(S1_under_prop)
df_S1_under_prop
```

```python
fig = plt.figure(figsize=(6,4))
fig.suptitle('S1 Electric underground proportion',fontsize=20)
plt.plot(df_S1_under_prop.index, df_S1_under_prop['el_underground_proportion'], 'yo-', label='Electric')
plt.xlabel('Analysis year',fontsize=14)
plt.ylabel('Underground proportion',fontsize=14)
plt.legend()
plt.savefig(r'../../figures/ST1-E-Underprop.png',bbox_inches='tight')
plt.show()
fig = plt.figure(figsize=(6,4))
fig.suptitle('S1 Broadband underground proportion',fontsize=20)
plt.plot(df_S1_under_prop.index, df_S1_under_prop['br_underground_proportion'], 'ro-', label='Broadband')
plt.xlabel('Analysis year',fontsize=14)
plt.ylabel('Underground proportion',fontsize=14)
plt.savefig(r'../../figures/ST1-B-Underprop.png',bbox_inches='tight')
plt.legend()
plt.show()
```
