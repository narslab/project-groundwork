```python
import pandas as pd
import numpy as np
import os
import sys
sys.path.append("../python/")
import csv
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches
import math
```

```python
### JAA sensitivity plots
# plotting cost sensitivity analysis results for SQ strategy
with open(r'../../results/outcomes/Cost/Sensitivity analysis/sensitivity-cost-result-JAA.csv') as ss:
    sensitivity_statusQuo=pd.read_csv(ss)
sns.set(rc={'figure.figsize':(14,7)})

# creating a dataframe
df = pd.DataFrame()
df['input'] = [sensitivity_statusQuo.iloc[0][0],sensitivity_statusQuo.iloc[1][0],sensitivity_statusQuo.iloc[2][0],sensitivity_statusQuo.iloc[3][0],sensitivity_statusQuo.iloc[4][0],sensitivity_statusQuo.iloc[5][0],sensitivity_statusQuo.iloc[6][0],sensitivity_statusQuo.iloc[7][0],sensitivity_statusQuo.iloc[8][0],sensitivity_statusQuo.iloc[9][0],sensitivity_statusQuo.iloc[10][0],sensitivity_statusQuo.iloc[11][0]]
df['+10%']=[sensitivity_statusQuo.iloc[0][1],sensitivity_statusQuo.iloc[1][1],sensitivity_statusQuo.iloc[2][1],sensitivity_statusQuo.iloc[3][1],sensitivity_statusQuo.iloc[4][1],sensitivity_statusQuo.iloc[5][1],sensitivity_statusQuo.iloc[6][1],sensitivity_statusQuo.iloc[7][1],sensitivity_statusQuo.iloc[8][1],sensitivity_statusQuo.iloc[9][1],sensitivity_statusQuo.iloc[10][1],sensitivity_statusQuo.iloc[11][1]]
df['-10%']=[sensitivity_statusQuo.iloc[0][2],sensitivity_statusQuo.iloc[1][2],sensitivity_statusQuo.iloc[2][2],sensitivity_statusQuo.iloc[3][2],sensitivity_statusQuo.iloc[4][2],sensitivity_statusQuo.iloc[5][2],sensitivity_statusQuo.iloc[6][2],sensitivity_statusQuo.iloc[7][2],sensitivity_statusQuo.iloc[8][2],sensitivity_statusQuo.iloc[9][2],sensitivity_statusQuo.iloc[10][2],sensitivity_statusQuo.iloc[11][2]]


#now stacking it
df2 = pd.melt(df, id_vars ='input', var_name='type of change', value_name='change in the output' )
sns.set(style='white')
fig, ax = plt.subplots()
for typ, df in zip(df2['type of change'].unique(),df2.groupby('type of change')):
    ax.barh(df[1]['input'], df[1]['change in the output'], height=0.3, label=typ)
ax.legend(title = 'Parameter change', fontsize=14)
ax.set_xlabel('Percentage change in costs', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
#ax.set_yticklabels(['$0$', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'])
ax.set_yticklabels([r'$c^{rep}_{E,O}$', r'$p^{oam}_{E,O}$',r'$p^{oam}_{E,U}$',r'$L_{E,U}$',r'$c^{rep}_{E,U}$', r'$w$',r'$L_{E,O}$', r'$a_i$', r'$a^{scl}$', r'$a^{shp}$', r'$r$', r'$p_{O,0}$',r'$\ell^{scl}$',r'$\ell^{shp}$'])
plt.grid()
plt.savefig(r'../../figures/cost-sensitivity-JAA.png', dpi=400, bbox_inches='tight')
```
