```python
import requests

def get_residential_buildings(api_key, city="Shrewsbury", state="MA"):
    # Define the ACS5 API endpoint for 2019 data
    url = "https://api.census.gov/data/2019/acs/acs5"
    # Specify the table ID for Units in Structure (verify the correct table ID)
    params = {
        "get": "NAME,B25024_001E,B25024_002E,B25024_003E,B25024_004E,B25024_005E,B25024_006E,B25024_007E,B25024_008E,B25024_009E",
        "for": "place:67675",  # Check this place code for Shrewsbury, MA
        "in": "state:25",       # Confirm state code for Massachusetts
        "key": api_key
    }

    # Make the API request
    response = requests.get(url, params=params)
    if response.status_code != 200:
        return f"Failed request with status code: {response.status_code}, Response text: {response.text}"

    try:
        data = response.json()
        return data[1:]  # Skip the header row in the response
    except ValueError:
        return "Error: No JSON data was found in the response"

# Replace 'your_api_key_here' with your actual Census API key
api_key = 'daf2bad631b83c1ceb4861102b13d8bc2bc2cc61'
residential_data = get_residential_buildings(api_key)
print(residential_data)
```

```python
import censusdata

# Function to get housing data for Shrewsbury, MA
def get_housing_data():
    # Specify the geographical area: Shrewsbury, Worcester County, Massachusetts
    geo = censusdata.censusgeo([('state', '25'), ('county', '027'), ('place', '60775')])  # Codes for Massachusetts, Worcester County, Shrewsbury
    
    # Fetch data: total housing units
    data = censusdata.download('acs5', 2019,
                                geo,
                                ['B25001_001E'])  # B25001_001E is the ACS variable for total housing units

    total_units = data.iloc[0]['B25001_001E']
    print(f"Total housing units in Shrewsbury, MA: {total_units}")

get_housing_data()
```
