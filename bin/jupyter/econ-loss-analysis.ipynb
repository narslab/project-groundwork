```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Define the function based on the given formulation
def SAI(tau, y, alpha, U_y):
    return tau * (alpha + (1 - U_y) * (1 - alpha))

# Constants
tau_0 = 1.38  # hours
alpha = 0.5  # Assumed alpha value, can be changed based on actual value

# U_y values from 0.34 to 1
U_y_values = np.linspace(0.34, 1, 100)
SAI_values = SAI(tau_0, 0, alpha, U_y_values)

# Plotting
plt.figure(dpi=300)
plt.plot(U_y_values, SAI_values, label='SAI_τ,y')

# Labels and title
plt.xlabel('$U_y$', fontsize=14)
plt.ylabel('$\\overline{SAI}_{\\tau, y}$ (hours)', fontsize=14)
plt.title('$\\overline{SAI}_{\\tau, y}$ vs $U_y$', fontsize=16)
plt.legend()
plt.grid(True)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Show plot
plt.show()
```
