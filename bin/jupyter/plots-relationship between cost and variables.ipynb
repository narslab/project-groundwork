```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# plotting statusquo strategy cost
with open(r'../../data/tidy/concord.csv') as concord:
    df_concord=pd.read_csv(concord)
    df_concord_variable=df_concord.filter(items=['total cost','dencity(#parcel/ft)','1 phase','SAND','CLAY','BLD_AREA/length','TOTAL_VAL/length'])
    #pd_concord_variable=pd_concord['total cost','dencity(#parcel/ft)','1 phase','SAND','CLAY','BLD_AREA','TOTAL_VAL']
pd.plotting.scatter_matrix(df_concord_variable, figsize  = [15, 15])
plt.figure(figsize=(12, 12))
```

```python
#plotting relationship between clay and sand
N=26
y = df_concord_variable['SAND']
x = df_concord_variable['CLAY']
colors = np.random.rand(N)
plt.title("relationship between SAND and CLAY")
plt.ylabel('SAND%')
plt.xlabel('CLAY%')
plt.legend(fontsize=16)
plt.scatter(x, y, c=colors, alpha=0.5)
plt.show()
```
