```python
import matplotlib.pyplot as plt
import Analyze_Result
import Simulate
import Network
from Analyze_Result import aggregate_through_years_statusQuo
from Analyze_Result import aggregate_through_years_under_after_lifespan
from Analyze_Result import calculate_additional_cost_from_under_after_lifespan
```

```python
###Calculating net present value of infrastructure, environmental, safety and total cost for statusQuo strategy
df_net_present_statusQuo=Analyze_Result.aggregate_through_years_statusQuo()
net_present_value_lifecycle_infrastructure_cost_statusQuo=[]
net_present_value_environmental_cost_statusQuo=[]
net_present_value_safety_cost_statusQuo=[]
net_present_value_total_statusQuo=[]
for index, row in df_net_present_statusQuo.iterrows():
    net_present_value_lifecycle_infrastructure_cost_statusQuo.append(row['total infra']/(1+Network.r)**index)
    net_present_value_environmental_cost_statusQuo.append(row['environmental restoration']/(1+Network.r)**index)
    net_present_value_safety_cost_statusQuo.append(row['total safety']/(1+Network.r)**index)
    net_present_value_total_statusQuo.append(row['total cost']/(1+Network.r)**index)
total_infrastructre_statusQuo=sum(net_present_value_lifecycle_infrastructure_cost_statusQuo)
total_environmental_statusQuo=sum(net_present_value_environmental_cost_statusQuo)
total_safety_statusQuo=sum(net_present_value_safety_cost_statusQuo)
total_total_statusQuo=sum(net_present_value_total_statusQuo)
```

```python
print(total_infrastructre_statusQuo)
print(total_environmental_statusQuo)
print(total_safety_statusQuo)
print(total_total_statusQuo)
```

```python
###plotting cost elements for statusQuo strategy
#fig = plt.figure()
plt.rcParams["figure.figsize"] = (12, 8)
plt.style.use('ggplot')

x = ['Lifecycle Infrastructure', 'Environmental Restoration', 'Safety', 'Total Cost']
cost = [total_infrastructre_statusQuo, total_environmental_statusQuo, total_safety_statusQuo, total_total_statusQuo]

x_pos = [i for i, _ in enumerate(x)]

plt.bar(x_pos, cost, color='green')

plt.xlabel("Cost Elements", fontsize=14)
plt.ylabel("Cost Output ($10millions)",fontsize=14)
plt.title("Net Preset Cost Output of statusQuo strategy", fontsize=20)

plt.xticks(x_pos, x)
plt.figsize=(28,25)
plt.show()
#savefig('test.png', bbox_inches='tight')
```
