```python
with open(r'../../results/outcomes/Cost/Simulation/S1-cost-simulation.csv') as S1_cost:
    df_S1_cost=pd.read_csv(S1_cost)
df_S1_cost=df_S1_cost.groupby(['year'])['capex_electric','opex_electric','environmental_restoration_electric','non_fatal_electric','fatal_electric','capex_broadband','opex_broadband'].sum()
df_S1_cost
```

```python
def plotStrategies(index, benefit=False, losses=False, cumulative=False):
    if benefit:
        with open(r'../../results/outcomes/Benefit/Simulation/S'+str(index)+'-benefit-simulation.csv') as s_file:
            df=pd.read_csv(s_file)
        if losses:
            df=df.groupby(['year'])['economic_losses_el', 'economic_loss_br'].sum() #'economic_losses_br', 'economic_losses_el'
        else:
            df=df.groupby(['year'])['aesthetic_benefit_el','aesthetic_benefit_br'].sum() #'economic_losses_br', 'economic_losses_el'
    else:
        with open(r'../../results/outcomes/Cost/Simulation/S'+str(index)+'-cost-simulation.csv') as s_file:
            df=pd.read_csv(s_file)
        df=df.groupby(['year'])['capex_el','opex_el','environmental_restoration_el','non_fatal_el','fatal_el','capex_br','opex_br'].sum()
    # plotting S2 cost results
    if cumulative:
        df=np.cumsum(df)
    df.plot.bar(stacked=True)
    fig = plt.gcf()
    fig.set_size_inches(15, 10)
    #plt.ylim([0, 50000000])
    plt.grid()
    if benefit:
        if losses:
            plt.title("S"+str(index)+" economic losses results", fontsize=30)
            plt.xlabel('Year',fontsize=25)
            plt.legend(['economic_losses_electric','economic_loss_broadband'], fontsize=20)
        else:
            plt.title("S"+str(index)+" aesthetic benefit results", fontsize=30)
            plt.xlabel('Year',fontsize=25)
            plt.legend(['aesthetic_benefit_electric','aesthetic_benefit_broadband'], fontsize=20)
    else:
        plt.title("S"+str(index)+" cost results", fontsize=30)
        plt.xlabel('Year',fontsize=25)
        plt.legend(['capex_electric','opex_electric','environmental_restoration_electric','non_fatal_electric','fatal_electric','capex_broadband','opex_broadband'], fontsize=20)
    plt.grid()
    plt.xticks(fontsize=18)
    plt.yticks(fontsize=18)
    if benefit:
        if losses:
            plt.savefig(r'../../figures/S'+str(index)+'_conomic_losses.png')
        else:
            plt.savefig(r'../../figures/S'+str(index)+'_aesthetic_benefit.png')
    else:
        plt.savefig(r'../../figures/S'+str(index)+'_cost.png')
```

```python
plotStrategies(13)
```

```python
strategies = [1,2,3,4,5,6,7,9,10]
for s in strategies:
    plotStrategies(s)
```
